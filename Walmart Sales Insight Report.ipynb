# Exploratory Data Analysis

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='State', data=df, palette='viridis')
plt.title('Count of Order ID by State')
plt.xlabel('State')
plt.ylabel('Count of Order ID')
plt.show()

```

```python
# Set labels and title
ax.set(xlabel='Country', ylabel='Count', title='Count of Orders by Country')
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='State', data=df, palette='viridis')
plt.title('Count of Sales by State')
plt.xlabel('State')
plt.ylabel('Count of Sales')
plt.show()
```

City

```python
# total number of orders from top 10 states

sales_state = df.groupby(['City'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'City',y= 'Sales')
```

# Quantity of orders by per state

```python
sales_state = df.groupby(['State'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Quantity')
```

# category of product

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sales_state = df.groupby(['Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Category',y= 'Sales')
```

        From above graphs we can see that most of the sold products are from Chairs, Phones and Tables category.

```python
sales_state = df.groupby(['Quantity'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Quantity',y= 'Sales')
```

```python
sales_state = df.groupby(['Sales'], as_index=False)['Order Date'].sum().sort_values(by='Order Date', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Sales',y= 'Order Date')
```

                                                   Sales by order dates

```python

sales_state = df.groupby(['State'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Profit')
```

                                                  Profit per state

```python
# total number of profit from top 10 cities

sales_state = df.groupby(['City'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'City',y= 'Profit')
```
